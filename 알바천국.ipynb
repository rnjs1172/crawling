```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
data = []
for i in range(1,11):
    url_1 = "http://www.alba.co.kr/story/albastory/StoryList.asp?page="+str(i)+"&pagesize=20&btypecd=101&schType=&schText=&schParam1=&schParam2=&boardType=1&OrderBy=regdt"
    html_1 = requests.get(url_1).text
    soup_1 = BeautifulSoup(html_1,"html.parser")
    for tag in soup_1.select("#Container .formList .title strong"):
        data.append(tag.text)
```
